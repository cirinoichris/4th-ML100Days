```python
# 資料篩選
HOUSETYPE_data = app_train[['TARGET', 'HOUSETYPE_MODE']]
```

```python
# 連續資料離散化
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

HOUSETYPE_data['HOUSETYPE_MODE_BINNED'] = le.fit_transform(HOUSETYPE_data['HOUSETYPE_MODE'])
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版

unique_house_type = app_train['HOUSETYPE_MODE'].unique()   # # 去重複

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    sns.distplot(app_train.loc[(app_train['HOUSETYPE_MODE_BINNED'] == unique_house_type[i]) & \
                              (app_train['TARGET'] == 0), 'HOUSETYPE_MODE'], label = "TARGET = 0", hist = False)
    sns.distplot(app_train.loc[(app_train['HOUSETYPE_MODE_BINNED'] == unique_house_type[i]) & \
                              (app_train['TARGET'] == 1), 'HOUSETYPE_MODE'], label = "TARGET = 1", hist = False) 
    plt.title(str(unique_house_type[i]))
plt.show()    
```
